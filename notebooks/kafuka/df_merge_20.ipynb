```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
citibikedemand = pd.read_csv('../Capstone_Project/cusp_capstone/data_for_notebooks/citi20_cleaned.csv',index_col=0)
```

```python
citibikedemand.head()
```

```python
citibikedemand.citi_inflow_counts.sum()
```

```python
citibikedemand['key'] = citibikedemand['month'].astype(str) +\
                        citibikedemand['day'].astype(str) +\
                        citibikedemand['night?'].astype(str) +\
                        citibikedemand['hex_id']
```

```python
mtademand = pd.read_csv('../Capstone_Project/cusp_capstone/data_for_notebooks/clena_df20_new.csv')
```

```python
mtademand.tail()
```

```python
mtademand['month']  = mtademand.apply(lambda row: row['date'].split('/')[0], axis=1)
mtademand['day']    = mtademand.apply(lambda row: row['date'].split('/')[1], axis=1)

mtademand['month']  = mtademand['month'].astype(int)
mtademand['day']    = mtademand['day'].astype(int)
```

```python
mtademand.shape
```

```python
mta20 = mtademand[mtademand['month']<5]
```

```python
mta20.shape
```

```python
hourgroups = {'8pm - 12am':1, 
              '12am - 4am':1, 
              '4am - 8am':0,
              '8am - 12pm':0,
              '12pm - 4pm':0,
              '4pm - 8pm':0}
```

```python
mta20['night?'] = mta20['timegroupstr'].map(hourgroups)
mta20['night?'] = mta20['night?'].astype(int)
```

```python
mta20.head(10)
```

```python
MTA_hex = pd.read_csv("../Capstone_Project/cusp_capstone/data_for_notebooks/mta_cleaned.csv")
```

```python
mta20_hex = pd.merge(mta20, MTA_hex, how='inner',left_on='c/a',right_on='Booth')
mta20_hex.head()
```

```python
# pivot table to sum day/night rides by hexagon, day, and month
mta20_cleaned = pd.pivot_table(mta20_hex,
                       values=['entries_diff', 'exit_diff'],
                       index=['hex_id_8', 'month', 'day', 'weekday', 'night?'],
                       aggfunc=np.sum).reset_index()
```

```python
mta20_cleaned = mta20_cleaned.rename(columns={"entries_diff": "mta_inflow_counts", "exit_diff": "mta_outflow_counts"})
```

```python
mta20_cleaned['key'] = mta20_cleaned['month'].astype(str) +\
                        mta20_cleaned['day'].astype(str) +\
                        mta20_cleaned['night?'].astype(str) +\
                        mta20_cleaned['hex_id_8']
```

```python
mtademand = mta20_cleaned[['mta_inflow_counts','mta_outflow_counts','key']]
```

```python
mtademand.head()
```

```python
mtademand.shape
```

```python
mtademand.mta_inflow_counts.sum()
```

```python
merged20 = citibikedemand.merge(mtademand, how ='left', on='key')
```
